```python
# ==========================================
# HANGMAN GAME
# Developed By: YASMIN FIRDOOS
# ==========================================

def display_word(secret_word, guessed_letters):

    word_display = ""

    for letter in secret_word:

        if letter in guessed_letters:
            word_display += letter + " "
        else:
            word_display += "_ "

    return word_display


def main():

    # Fixed word for easy testing and screenshots
    secret_word = "python"

    guessed_letters = []
    attempts = 6

    print("=" * 50)
    print("              HANGMAN GAME")
    print("=" * 50)

    while attempts > 0:

        current_word = display_word(
            secret_word,
            guessed_letters
        )

        print("\nWord:", current_word)

        # Win condition
        if "_" not in current_word:

            print("\n🎉 Congratulations!")
            print("You guessed the word successfully.")
            break

        guess = input("Enter a letter: ").lower().strip()

        # Input validation
        if len(guess) != 1 or not guess.isalpha():

            print("❌ Enter only one alphabet letter.")
            continue

        # Repeated letter check
        if guess in guessed_letters:

            print("⚠ Letter already guessed.")
            continue

        guessed_letters.append(guess)

        if guess in secret_word:

            print("✅ Correct Guess!")

        else:

            attempts -= 1

            print("❌ Wrong Guess!")
            print("Attempts Left:", attempts)

    if attempts == 0:

        print("\n💀 Game Over!")
        print("The word was:", secret_word)

    print("\n✅ Thank You For Playing!")


if __name__ == "__main__":
    main()
```

```python
# ==========================================
# STOCK PORTFOLIO TRACKER
# Developed By: Yasmin Firdoos Shaik
# ==========================================

stock_prices = {
    "AAPL": 180,       # Apple
    "TSLA": 250,       # Tesla
    "GOOG": 150,       # Google
    "MSFT": 320,       # Microsoft
    "AMZN": 140,       # Amazon
    "CODEALPHA": 500   # Custom Internship Stock
}


def get_portfolio():

    portfolio = {}

    while True:

        stock = input(
            "\nEnter Stock Symbol "
            "(AAPL/TSLA/GOOG/MSFT/AMZN/CODEALPHA)\n"
            "Type DONE to finish: "
        ).upper().strip()

        if stock == "DONE":
            break

        if stock not in stock_prices:
            print("❌ Invalid Stock Symbol!")
            continue

        try:

            quantity = int(input("Enter Quantity: "))

            if quantity <= 0:
                print("❌ Quantity must be greater than zero.")
                continue

        except ValueError:

            print("❌ Please enter a valid quantity.")
            continue

        portfolio[stock] = portfolio.get(stock, 0) + quantity

    return portfolio


def calculate_portfolio(portfolio):

    total_value = 0
    report_data = []

    for stock, quantity in portfolio.items():

        price = stock_prices[stock]
        investment = price * quantity

        total_value += investment

        report_data.append(
            (stock, quantity, price, investment)
        )

    return total_value, report_data


def display_summary(report_data, total_value):

    print("\n" + "=" * 50)
    print("            PORTFOLIO SUMMARY")
    print("=" * 50)

    for stock, quantity, price, investment in report_data:

        print(
            f"Stock: {stock} | "
            f"Quantity: {quantity} | "
            f"Price: ${price} | "
            f"Investment Value: ${investment}"
        )

    print("=" * 50)
    print(f"Total Portfolio Value = ${total_value}")
    print("=" * 50)


def save_report(report_data, total_value):

    with open("portfolio_report.txt", "w") as file:

        file.write("CODEALPHA STOCK PORTFOLIO REPORT\n")
        file.write("=" * 50 + "\n\n")

        for stock, quantity, price, investment in report_data:

            file.write(
                f"Stock: {stock} | "
                f"Quantity: {quantity} | "
                f"Price: ${price} | "
                f"Investment Value: ${investment}\n"
            )

        file.write("\n" + "=" * 50 + "\n")
        file.write(f"Total Portfolio Value = ${total_value}\n")
        file.write("=" * 50)

    print("\n✅ Portfolio report saved as 'portfolio_report.txt'")


def main():

    print("=" * 50)
    print("   CODEALPHA STOCK PORTFOLIO TRACKER")
    print("=" * 50)
    print("Internship Project - CodeAlpha")
    print("Developed By: Yasmin Firdoos Shaik")

    portfolio = get_portfolio()

    if not portfolio:
        print("\n❌ No stocks entered.")
        return

    total_value, report_data = calculate_portfolio(portfolio)

    display_summary(report_data, total_value)

    save_report(report_data, total_value)

    print("\n✅ Thank You For Using Stock Portfolio Tracker!")
    print("✅ Internship Project Completed Successfully.")


if __name__ == "__main__":
    main()
```

```python
# ==========================================
# SMART RULE-BASED CHATBOT
#Developed By: yasmin firdoos shaik
# ==========================================


def get_response(user_message):
    """
    Response Processing Module

    Maps user queries to predefined responses
    using keyword matching techniques.
    """

    responses = {
        "hello": "Hi! How can I help you?",
        "hi": "Hello! Nice to meet you.",
        "how are you": "I am fine. Thank you for asking.",
        "what is your name": "My name is Python Chatbot.",
        "who created you": "I was created using Python.",
        "python": "Python is a popular programming language used for AI, Automation, and Web Development.",
        "codealpha": "CodeAlpha provides internship opportunities to enhance practical programming skills.",
        "project": "This chatbot is a Rule-Based Conversational AI System.",
        "thank you": "You're welcome!",
        "bye": "Goodbye! Have a great day."
    }

    # Pattern Matching Logic
    for keyword, response in responses.items():

        if keyword in user_message:
            return response

    return (
        "Sorry, I don't understand that. "
        "Please try another question."
    )


def chatbot():
    """
    Main Interaction Engine

    Handles user communication,
    response generation,
    and session management.
    """

    print("=" * 50)
    print("      SMART RULE-BASED CHATBOT")
    print("=" * 50)

    print("\nInternship Project - CodeAlpha")
    print("Type 'bye' anytime to exit.\n")

    while True:

        # User Input Processing
        user_message = input("You: ").lower().strip()

        # Input Validation
        if not user_message:

            print("Bot: Please enter a message.")
            continue

        # Generate Response
        bot_response = get_response(user_message)

        print("Bot:", bot_response)

        # Session Termination Condition
        if "bye" in user_message:

            print("\n✅ Chatbot Session Closed Successfully.")
            break


def main():
    """
    Main Program Controller
    """

    chatbot()


if __name__ == "__main__":
    main()
```
